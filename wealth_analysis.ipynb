```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bondholders = pd.read_csv("data/Bondholders and the Second bank - Sheet1.csv")
sns.set_style("whitegrid") 
# Change color palette
sns.set_palette("pastel")
sns.set_context("notebook")  # Adjust according to the intended use
```

```python
list_of_banks = bondholders[bondholders["Bond"] == "six percent loan (1812)"][["First Name", "Amount", "State", "Town"]] 
list_of_banks = list_of_banks.sort_values(by="Amount", ascending=False)
ax = list_of_banks.plot.bar(x="First Name", y="Amount", rot=90) 
ax.set_title("Amount of Subscriptions Held By Each Bank (Six Percent Loan of 1812)")
ax.set_xlabel("Bank Name")
ax.set_ylabel("Amount (dollars)")
ax.figure.savefig("subscriptions.png", dpi=300)
```

```python
total_amt = 4200000
print(list_of_banks.head())
list_of_banks["First Name"] = list_of_banks["First Name"] + " (" + list_of_banks["State"] + ") "  
list_of_banks["percent_amt"] = list_of_banks["Amount"] / total_amt 
ax = list_of_banks.plot.bar(x="First Name", y="percent_amt", rot=90) 
ax.grid(False)
ax.set_title("Percentage of Subscriptions Held By Each Bank (Six Percent Loan of 1812)", pad=20)
ax.set_xlabel("Bank Name")
ax.set_ylabel("Percentage of Total Subscriptions", labelpad=20)
ax.get_legend().remove()

ax2 = ax.twinx()
ax2.set_ylim(list_of_banks["Amount"].min(), list_of_banks["Amount"].max())
ax2.grid(False)
ax2.set_ylabel("Subscription Amount", labelpad=20)

ax.figure.savefig("percentages.png", dpi=300, bbox_inches='tight')
```

```python
bank_count = int(bondholders["Type"].value_counts().loc["Bank"])
individual_count = int(bondholders["Type"].value_counts().loc["Individual"])
corporate_count = int(bondholders["Type"].value_counts().loc["Corporate entry"])
```

```python
# Number of banks versus indiviudals 
# num_banks = bondholders.plot.pie(y="Type", figsize=(5, 5), labels=["Individuals", "Banks"])
plt.pie(x=[bank_count, individual_count, corporate_count], labels=["Bank", "Individuals", "Corporate Entry"], autopct="%1.1f%%")
plt.title("Proportion of Banks versus Individuals")
plt.savefig("proportion_banks_versus_individuals")
plt.show()
```

```python
# Amount by banks versus individuals - six percent loan of 1812 
plt.pie(x=[4190000, 1928900], labels=["Bank", "Individuals"], autopct="%1.1f%%")
plt.title("Amount of Eleven Million Loan (1812) Subscribed by Banks versus Individuals (May)")
plt.savefig("six_percent_loan_banks.png")
plt.show()
```

```python
# Amount owned by individuals and banks - Sixteen Million Loan of 1813 
total = 16000000
girard = 3100000
parish = 3100000
astor = 1500000
junior_members = 2400000
sizes = [girard, parish, astor, junior_members, total-(girard+parish+astor+junior_members)]

# Function to format the autopct
def autopct_format(pct, raw_data):
    absolute = pct/100.*sum(raw_data)/1.0e6
    #absolute = int(round(pct/100.*sum(raw_data)))
    return f"{pct:.1f}% (${absolute:.1f}m)"

plt.pie(x=sizes, labels=["Girard", "Parish", "Astor", "Junior Members", "Other"], autopct=lambda pct: autopct_format(pct, sizes))
plt.savefig("sixteen_million_loan.png", dpi=300, bbox_inches='tight')
plt.show()
```
